# Univariate Analysis (non visualization : numeriacal)

```python
def univariate_analysis(numerical_data):
    for column in numerical_data.columns:
        print("*" * 5, column, "*" * 5)
        print("min:", numerical_data[column].min())
        print("max:", numerical_data[column].max())
        print("mean:", numerical_data[column].mean())
        print("median:", numerical_data[column].median())
        print("mode:", numerical_data[column].mode()[0])  # Get the first mode
        print("skewness:", numerical_data[column].skew())
        print("kurtosis:", numerical_data[column].kurt())
        print("---------------------------------------------")

# Assuming 'df' is your DataFrame, define the numerical columns
numerical_columns = Data[["Salary", "10percentage", "12percentage", "collegeGPA", "Domain", 
                         "conscientiousness", "agreeableness", "extraversion", "openess_to_experience"]]

# Call the function
univariate_analysis(numerical_columns)
```

#Observation: 
1.Salary shows extreme positive skewness, with most salaries clustered around ₹3,00,000 but a few very high outliers pulling the mean upward.
2.10th and 12th Percentages** are moderately high on average, with slight negative skewness, meaning most students performed well, with few lower scores.
3.Personality Traits like conscientiousness, agreeableness, and openness to experience are generally negative or neutral, with some traits showing moderate skewness toward lower values.

# Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
import scipy.stats as stats

# Creating the crosstab
crosstab = pd.crosstab(Data["Specialization"], Data["Gender"])

# Perform the Chi-Square test of independence
chi2_stat, p_value, dof, expected = stats.chi2_contingency(crosstab)

# Display results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant relationship between Gender and Specialization.")
else:
    print("Fail to reject the null hypothesis: No significant relationship between Gender and Specialization.")
```

The Chi-Square test reveals a significant relationship between gender and specialization, as the p-value is far below 0.05. This indicates that specialization preferences vary significantly between male and female students.
